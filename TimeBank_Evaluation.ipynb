```python
def process_xml(filepath):
  try:
    mytree = ET.parse(filepath)

    root = mytree.getroot()
    timex3s = root.findall('TIMEX3')
    timex3s = [t.text for t in timex3s]
    MIs = root.findall('MAKEINSTANCE')
    signals = root.findall('SIGNAL')
    signals = [s.text for s in signals]
    tlinks = root.findall('TLINK')
    tlinks = [t.attrib for t in tlinks]
    mi_attribs = []
    mi_texts = []
    for mi in MIs:
      mi_attribs.append(mi.attrib)
      mi_texts.append(mi.text)

    rough_string = ET.tostring(root, 'utf-8')
    reparsed = minidom.parseString(rough_string)
    xml_str = reparsed.toprettyxml(indent ="\t") 
    return xml_str, timex3s, signals, tlinks, mi_texts, mi_attribs
  except:
    print(filepath)
  
```

```python
df = pd.DataFrame({'ID': filenames})
```

```python
elems = ['Stanza', 'SpaCy', 'Natasha', 'manual']


for e in elems:
  xml_strs, timex3s, signals, tlinks, mi_texts, mi_attribs = list(), list(), list(), list(), list(), list()

  for f in filenames:
    process_xml(path + f'{e}/{f}')
    xml_str, timex3, signal, tlink, mi_text, mi_attrib = process_xml(path + f'{e}/{f}')
    xml_strs.append(xml_str)
    timex3s.append(timex3)
    signals.append(signal)
    tlinks.append(tlink)
    mi_texts.append(mi_text)
    mi_attribs.append(mi_attrib)

  df[f'xml{e}'] = xml_strs
  df[f'timex3s{e}'] = timex3s
  df[f'signals{e}'] = signals
  df[f'tlinks{e}'] = tlinks
  df[f'MItexts{e}'] = mi_texts
  df[f'MIattributes{e}'] = mi_attribs
```
