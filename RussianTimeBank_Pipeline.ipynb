```python
# Extract Named Entities from an XML-file, returns a list of NEs

def extract_NEs(file):
  mytree = ET.parse(file)
  root = mytree.getroot()
  text = root.text
  doc = nlp_s(text)
  dEnts = doc.ents
  out = []
  for de in dEnts:
    out.append((de.text, de.start_char, de.end_char, de.type))
  return out
```

```python
def get_roots_stanza(text):
  doc = nlp_s(text)
  predicates = ""
  for sent in doc.sentences:
    for word in sent.words:
      deprels = ['root', 'acl', 'advcl', 'advmod']
      if word.deprel in deprels and word.upos == 'VERB':
        feats = word.feats
        tenses = ['Past', 'Pres']
        if re.search('Tense=(.+?)\|', feats):
          if re.search('Tense=(.+?)\|', feats).group(1) in tenses:
              predicates += f'{word.text},{word.start_char},{word.end_char};'

            # predicates.append((word.text, word.start_char, word.end_char))

  return predicates
```

```python
df['PredicatesStanza'] = df['Text'].progress_apply(get_roots_stanza)
```

```python
def get_roots_spacy(text):
  doc = nlp(text)
  predicates = ""
  for token in doc:
    deprels = ['ROOT', 'acl', 'advcl', 'advmod']
    if token.dep_ in deprels and token.pos_ == 'VERB':
      tenses = ['Pres', 'Past']
      if token.morph.get("Tense"):
        tense = token.morph.get("Tense")[0]
        if tense in tenses:
          predicates += f'{token.text},{token.idx},{token.idx + len(token.text)};'
  return predicates
```

```python
df['PredicatesSpaCy'] = df['Text'].progress_apply(get_roots_spacy)
```

### Natasha

```python
def get_roots_natasha(text):
  doc = Doc(text)
  doc.segment(segmenter)
  doc.parse_syntax(syntax_parser)
  doc.tag_morph(morph_tagger)
  predicates = ""
  for t in doc.tokens:
    deprels = ['root', 'acl', 'advcl', 'advmod']
    if t.rel in deprels and t.pos == 'VERB':
      tenses = ['Past', 'Pres']
      if 'Tense' in t.feats.keys():
        tense = t.feats['Tense']
        if tense in tenses:
          predicates += f'{t.text},{t.start},{t.stop};'
  return predicates
```

```python
def add_dct(filepath, id):
  mytree = ET.parse(filepath)
  root = mytree.getroot()
  
  formatted_dct = f'{id[:4]}/{id[4:6]}/{id[6:8]}'

  timex3 = ET.SubElement(root, 'TIMEX3')
  timex3.set('tid', 't1')
  timex3.set('type', 'DATE')
  timex3.set('value', formatted_dct)
  timex3.set('temporalFunction', 'false')
  timex3.set('functionInDocument', 'PUBLICATION_TIME')

  timex3.text = formatted_dct
  rough_string = ET.tostring(root, 'utf-8')
  reparsed = minidom.parseString(rough_string)
  xml_str = reparsed.toprettyxml(indent ="\t") 
  return xml_str
```

```python
def get_attributes_stanza(text, predicate):
  pred_text, pred_start, pred_end = predicate.split(',')
  doc = nlp_s(text)

  for sent in doc.sentences:
    for token in sent.tokens:
      if token.start_char == int(pred_start) and token.end_char == int(pred_end):
        word = token.words[0]
        word_id = word.id
        TEXT = word.text
        POS = word.upos
        features = word.feats
        TENSE = 'NONE'
        ASPECT = 'NONE'
        MOOD = 'NONE'
        POLARITY = None

        if re.search('Tense=(.+?)\|', features):
          TENSE = re.search('Tense=(.+?)\|', features).group(1)

        if re.search('Aspect=(.+?)\|', features):
          ASPECT = re.search('Aspect=(.+?)\|', features).group(1)

        if re.search('Mood=(.+?)\|', features):
          MOOD = re.search('Mood=(.+?)\|', features).group(1)


        negations = [] 
        for word in sent.words:
          if word.text.lower() == 'не' and word.upos == 'PART':
            negations.append(word)

        for n in negations:
          if n.head == word_id:
            POLARITY = 'NEG'

        return POS, TENSE, ASPECT, MOOD, POLARITY, TEXT
```

```python
def get_attributes_spacy(text, predicate):
  pred_text, pred_start, pred_end = predicate.split(',')
  doc = nlp(text)
  for sent in doc.sents:
    docu = nlp(str(sent))
    for token in docu:
      if token.text == pred_text:
        TEXT = token.text
        POS = token.pos_

        TENSE = 'NONE'
        ASPECT = 'NONE'
        MOOD = 'NONE'
        POLARITY = None

        if token.morph.get('Tense'):
          TENSE = token.morph.get('Tense')[0]
        if token.morph.get('Aspect'):
          ASPECT = token.morph.get('Aspect')[0]
        if token.morph.get('Mood'):
          MOOD = token.morph.get('Mood')[0]

        for ch in token.children:
          if ch.text.lower() == 'не' and ch.pos_ == 'PART':
            POLARITY = 'NEG'
        
        return POS, TENSE, ASPECT, MOOD, POLARITY, TEXT
        
```

```python
def get_attributes_natasha(text, predicate):
  pred_text, pred_start, pred_end = predicate.split(',')

  doc = Doc(text)
  doc.segment(segmenter)
  doc.parse_syntax(syntax_parser)
  doc.tag_morph(morph_tagger)  

  for sent in doc.sents:
    for token in sent.tokens:
      if token.start == int(pred_start) and token.stop == int(pred_end):
        TEXT = token.text
        POS = token.pos
        TENSE = 'NONE'
        ASPECT = 'NONE'
        MOOD = 'NONE'
        POLARITY = None
        features = token.feats

        if 'Tense' in features.keys():
          TENSE = features['Tense']
        if 'Aspect' in features.keys():
          ASPECT = features['Aspect']
        if 'Mood' in features.keys():
          MOOD = features['Mood']

        negations = []
        for t in sent.tokens:
          if t.text.lower == 'не' and t.pos == 'PART':
            if t.head_id == token.id:
              POLARITY = 'NEG'
  
        return POS, TENSE, ASPECT, MOOD, POLARITY, TEXT
```

```python
def add_instances(filepath, text, model, events):
  mytree = ET.parse(filepath)
  root = mytree.getroot()

  #add tags for event instances:
  id_counter = 1
  if model == 'Stanza':
    for e in events:
      POS, TENSE, ASPECT, MOOD, POLARITY, TEXT = get_attributes_stanza(text, e)

      ME = ET.SubElement(root, 'MAKEINSTANCE')
      ME.set('eiid', f'ei{id_counter}')
      ME.set('pos', POS)
      ME.set('tense', TENSE)
      ME.set('aspect', ASPECT)
      ME.set('modality', MOOD)
      # ME.set('cardinality', '???') -- only present if there is a signal id that suggests cardinality
      
      if POLARITY:
        ME.set('polarity', POLARITY)
        
      ME.text = TEXT  
      id_counter += 1
  
  elif model == 'SpaCy':
    for e in events:
      POS, TENSE, ASPECT, MOOD, POLARITY, TEXT = get_attributes_spacy(text, e)

      ME = ET.SubElement(root, 'MAKEINSTANCE')
      ME.set('eiid', f'ei{id_counter}')
      ME.set('pos', POS)
      ME.set('tense', TENSE)
      ME.set('aspect', ASPECT)
      ME.set('modality', MOOD)
      
      # ME.set('cardinality', '???') -- only present if there is a signal id that suggests cardinality
      
      if POLARITY:
        ME.set('polarity', POLARITY)
        
      ME.text = TEXT

      id_counter += 1

  elif model == 'Natasha':
    for e in events:
      POS, TENSE, ASPECT, MOOD, POLARITY, TEXT = get_attributes_natasha(text, e)

      ME = ET.SubElement(root, 'MAKEINSTANCE')
      ME.set('eiid', f'ei{id_counter}')
      ME.set('pos', POS)
      ME.set('tense', TENSE)
      ME.set('aspect', ASPECT)
      ME.set('modality', MOOD)
      # ME.set('cardinality', '???') -- only present if there is a signal id that suggests cardinality
      
      if POLARITY:
        ME.set('polarity', POLARITY)
        
      ME.text = TEXT

      id_counter += 1    

  rough_string = ET.tostring(root, 'utf-8')
  reparsed = minidom.parseString(rough_string)
  xml_str = reparsed.toprettyxml(indent ="\t") 
  
  return xml_str
```

```python
def detect_signals_spacy(text, filepath):
  signals = ['в', 'с', 'до', 'через', 'на', 'по', 'к', 'около', 'под', 'за']

  doc = nlp(text)

  mytree = ET.parse(filepath)
  root = mytree.getroot()

  # get instances of verbs and temporal expressions to link to signals:
  instances = root.iter('MAKEINSTANCE')
  ins_text = [i.text for i in instances]

  timex3s = root.findall('TIMEX3')

  times_text = [i.text for i in timex3s]
  # only selecting the first word of TIMEX3 if it's a phrase:
  times_text_ = [i.split(' ')[0] if len(i.split(' ')) > 1 else i for i in times_text]
  times_ids = [i.attrib['tid'] for i in timex3s]

  needed_tids = []
  needed_signals = []
  instances_to_link = []
  counter = 1

  # iterate through dependencies:
  for sent in doc.sents:
    for t in range(len(sent)):
      for tt in range(len(times_text_)):
        if sent[t].text == times_text_[tt]:
          if sent[t-1].text.lower() in signals:
            needed_tids.append(times_ids[tt])
            needed_signals.append(sent[t-1].text.lower())

            difference = len(times_text[tt].split(' ')) - len(times_text_[tt].split(' '))

            if difference > 0:
              for d in range(difference + 1):
                if sent[t+d].head.text in ins_text:
                  instances_to_link.append((sent[t+d].head.text, 's' + str(counter)))
                  counter += 1
            else:
              head = sent[t].head.text
              if head in ins_text:
                instances_to_link.append((sent[t].head.text, 's' + str(counter)))
                counter += 1
            

  for itl in instances_to_link:
    for i in root.iter('MAKEINSTANCE'):
      if i.text == itl[0]:
        i.set('SignalID', itl[1])

  rough_string = ET.tostring(root, 'utf-8')
  reparsed = minidom.parseString(rough_string)
  xml_str = reparsed.toprettyxml(indent ="\t") 

  sid = 1
  # add SIGNAL tags:
  for nt in range(len(needed_tids)):
    regex_for_insert = f'(\w)+\s+?<TIMEX3 tid="{needed_tids[nt]}'
    res = re.search(regex_for_insert, xml_str)
    if res:
      start_index = res.start()
      end_index = start_index + len(needed_signals[nt])
      xml_str = xml_str[:start_index] + f'<SIGNAL sid="s{sid}">' + needed_signals[nt] + '</SIGNAL>' + xml_str[end_index:]
      sid += 1
  
  return xml_str
```

```python
def detect_signals_natasha(text, filepath):
  signals = ['в', 'с', 'до', 'через', 'на', 'по', 'к', 'около', 'под', 'за']

  doc = Doc(text)
  doc.segment(segmenter)
  doc.parse_syntax(syntax_parser)
  doc.tag_morph(morph_tagger)

  mytree = ET.parse(filepath)
  root = mytree.getroot()

  # get instances of verbs and temporal expressions to link to signals:
  instances = root.iter('MAKEINSTANCE')
  ins_text = [i.text for i in instances]

  timex3s = root.findall('TIMEX3')

  times_text = [i.text for i in timex3s]
  # only selecting the first word of TIMEX3 if it's a phrase:
  times_text_ = [i.split(' ')[0] if len(i.split(' ')) > 1 else i for i in times_text]
  times_ids = [i.attrib['tid'] for i in timex3s]

  needed_tids = []
  needed_signals = []
  instances_to_link = []
  counter = 1
  for sent in doc.sents:
    for t in range(len(sent.tokens)):
      for tt in range(len(times_text_)):
        if sent.tokens[t].text == times_text_[tt]: # 
          if sent.tokens[t-1].text.lower() in signals:

            needed_tids.append(times_ids[tt])
            needed_signals.append(sent.tokens[t-1].text.lower())

            difference = len(times_text[tt].split(' ')) - len(times_text_[tt].split(' '))

            if difference > 0:
              for d in range(difference + 1):
                head_id = sent.tokens[t+d].head_id
                
                for pot_head in sent.tokens:
                  if pot_head.id == head_id:
                    head_text = pot_head.text

                if head_text in ins_text:
                  instances_to_link.append((head_text, 's' + str(counter)))
                  counter += 1
            else:
              head_id = sent.tokens[t].head_id
                
              for pot_head in sent.tokens:
                if pot_head.id == head_id:
                  head_text = pot_head.text
              
              if head_text in ins_text:
                instances_to_link.append((head_text, 's' + str(counter)))
                counter += 1
            

  for itl in instances_to_link:
    for i in root.iter('MAKEINSTANCE'):
      if i.text == itl[0]:
        i.set('SignalID', itl[1])

  rough_string = ET.tostring(root, 'utf-8')
  reparsed = minidom.parseString(rough_string)
  xml_str = reparsed.toprettyxml(indent ="\t") 

  sid = 1
  # add SIGNAL tags:
  for nt in range(len(needed_tids)):
    regex_for_insert = f'(\w)+\s+?<TIMEX3 tid="{needed_tids[nt]}'
    res = re.search(regex_for_insert, xml_str)
    if res:
      start_index = res.start()
      end_index = start_index + len(needed_signals[nt])
      xml_str = xml_str[:start_index] + f'<SIGNAL sid="s{sid}">' + needed_signals[nt] + '</SIGNAL>' + xml_str[end_index:]
      sid += 1
  
  return xml_str
```

```python
def detect_links_spacy(text, filepath):

  doc = nlp(text)

  mytree = ET.parse(filepath)
  root = mytree.getroot()

  timex3s = root.findall('TIMEX3')

  times_text = [i.text for i in timex3s]
  # only selecting the first word of TIMEX3 if it's a phrase:
  times_text_ = [i.split(' ')[0] if len(i.split(' ')) > 1 else i for i in times_text]
  times_ids = [i.attrib['tid'] for i in timex3s]

  tlinks = []

  # iterate through dependencies:
  for sent in doc.sents:
    for t in range(len(sent)):
      for tt in range(len(times_text_)):
        if sent[t].text == times_text_[tt]:
          # check if the TE has a dep relation to another TE
          tokens1 = times_text[tt].split(' ')
          difference = len(tokens1) - len(times_text_[tt].split(' '))

          if difference > 0:
            if sent[t+1].text == tokens1[1]:

              for other_t in range(len(sent)):
                if sent[other_t].text in times_text_:
                  token2_id = times_text_.index(sent[other_t].text)
                  tokens2 = times_text[token2_id].split(' ')
                  difference_2 = len(tokens2) - 1
                  if difference_2 > 0:
                    for d2 in range(difference_2 + 1):
                      for d in range(difference + 1):
                        token1 = tokens1[d]
                        token2 = tokens2[d2]
                        for ttt in range(len(sent)):
                          if sent[ttt].text == token1 and sent[ttt].head.text == token2:
                            tid1_index = times_text_.index(tokens1[0])
                            tid1 = times_ids[tid1_index]
                            tid2_index = times_text_.index(tokens2[0])
                            tid2 = times_ids[tid2_index]

                            tlinks.append((tid1, tid2))

                  else:
                    for d in range(difference + 1):
                      token1 = tokens1[d]
                      tid1_index = times_text_.index(tokens1[0])
                      tid1 = times_ids[tid1_index]
                      tid2_index = times_text_.index(sent[other_t].text)
                      tid2 = times_ids[tid2_index]
                      token2 = sent[other_t].text
                      if sent[t].head.text == sent[other_t].text:
                        tlinks.append((tid1, tid2))


          else:
            for other_t in range(len(sent)):
              if sent[other_t].text in times_text_:
                exp_index = times_text_.index(sent[other_t].text)
                tokens2 = times_text[exp_index].split(' ')
                difference = len(sent[other_t].text.split(' ')) - len(sent[other_t].text.split(' '))
                if difference > 0:
                  for d in range(difference + 1):
                    token2 = tokens2[d]
                    if sent[t].head.text == token2:
                      other_id = times_text_.index(sent[t].head.text)
                      tlinks.append((times_ids[other_id], times_ids[tt]))

                else:
                  if sent[t].head.text == other_t:
                    other_id = times_text_.index(other_t.head.text)
                    tlinks.append((times_ids[other_id], times_ids[tt]))



  tlinks = list(set(tlinks))

  for tl in tlinks:
    if tl[0] != tl[1]:
      tlink = ET.SubElement(root, 'TLINK')
      tlink.set('timeID', tl[0])
      tlink.set('relatedToTime', tl[1])


  rough_string = ET.tostring(root, 'utf-8')
  reparsed = minidom.parseString(rough_string)
  xml_str = reparsed.toprettyxml(indent ="\t") 

  
  return xml_str
```
